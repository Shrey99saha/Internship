Q2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
url_2=requests.get('https://www.imdb.com/chart/top/')
url_2
```

```python
soup=BeautifulSoup(url_2.content)
```

```python
Movies = []
for i in soup.find_all('td',class_="titleColumn"):
    Movies.append(i.get_text().strip(" ").replace('\n',"  "))
Movies
```

```python
Year = []
for i in soup.find_all('span',class_="secondaryInfo"): 
    Year.append(i.text)
Year
```

```python
ratings = []
for i in soup.find_all('td',class_="ratingColumn imdbRating"): 
    ratings.append(i.text)
ratings
```

```python
data=pd.DataFrame()
data["Movie Name"]=Movies
data["Year Of Release"]=Year
data["Ratings"]=ratings
data.to_csv("IMDB’s Top rated movies",index=False)
```

```python
pd.read_csv("IMDB’s Top rated movies")
```

Q9. Write a python program to scrape mentioned details from dineout.co.in

```python
url_9=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
url_9
```

```python
soup=BeautifulSoup(url_9.content)
```

```python
Resturants = []
for i in soup.find_all('a',class_="restnt-name ellipsis"): 
    Resturants.append(i.text)
Resturants
```

```python
Cuisines = []
for i in soup.find_all('span',class_="double-line-ellipsis"): 
    Cuisines.append(i.text.split('|')[1])
Cuisines
```

```python
Location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"): 
    Location.append(i.text)
Location
```

```python
Ratings = []
for i in soup.find_all('div',class_="restnt-rating rating-4"): 
    Ratings.append(i.text)
Ratings
```

```python
img = []
for i in soup.find_all('img',class_="no-img"): 
    img.append(i.get('data-src'))
img
```
